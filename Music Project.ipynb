```python
import pandas as pd
song = pd.read_csv("SongList.csv")
song
```

```python
import plotly.graph_objs as go
import pandas as pd

# Count movies per actor
movies_per_actor = song.groupby('Actor')['Movie'].nunique().sort_values(ascending=False)

# Select top 15 actors by number of movies
top_actors = movies_per_actor.head(15)

# Create a Plotly bar chart with annotations
data_movies_per_actor = go.Bar(
    x=top_actors.index,
    y=top_actors.values,
    marker=dict(color='skyblue'),
    text=top_actors.values,
    textposition='auto',
)

layout_movies_per_actor = go.Layout(
    title='Top 15 Actors by Number of Movies',
    xaxis=dict(title='Actor'),
    yaxis=dict(title='Number of Movies'),
    height=800,
)

fig_movies_per_actor = go.Figure(data=[data_movies_per_actor], layout=layout_movies_per_actor)

# Update layout for aesthetic improvements
fig_movies_per_actor.update_traces(textfont_size=12, textfont_color='black')  # Text font size and color
fig_movies_per_actor.update_layout(plot_bgcolor='rgba(0,0,0,0)',  # Transparent plot background
                                   paper_bgcolor='rgba(0,0,0,0)',  # Transparent paper background
                                   font=dict(family="Arial, sans-serif", size=14, color='darkslategray'),  # Font settings
                                   title=dict(x=0.5, y=0.95, font=dict(size=24)),  # Title position and style
                                   xaxis=dict(tickfont=dict(size=12)),  # X-axis tick font size
                                   yaxis=dict(tickfont=dict(size=12)),  # Y-axis tick font size
                                   )

# Display plot
fig_movies_per_actor.show()

```

```python
import pandas as pd

# Assuming 'Year' column is in integer format
song['Decade'] = song['Year'] // 10 * 10  # Create a new column for decade

# Aggregate number of songs per decade
songs_per_decade = song.groupby('Decade')['Songs'].count()

# Aggregate number of unique music directors per decade
unique_music_directors_per_decade = song.groupby('Decade')['Music Director'].nunique()

# Aggregate number of unique directors per decade
unique_directors_per_decade = song.groupby('Decade')['Director'].nunique()

# Aggregate number of unique actors per decade
unique_actors_per_decade = song.groupby('Decade')['Actor'].nunique()

# Combine all aggregated data into a DataFrame
decade_stats = pd.DataFrame({
    'Songs': songs_per_decade,
    'Unique Music Directors': unique_music_directors_per_decade,
    'Unique Directors': unique_directors_per_decade,
    'Unique Actors': unique_actors_per_decade
})

# Print aggregated statistics
print("Decade-wise Statistics:")
print(decade_stats)
```

```python
import plotly.graph_objs as go

# Create a Plotly bar chart for songs per decade
data_songs_per_decade = go.Bar(
    x=decade_stats.index,
    y=decade_stats['Songs'],
    marker=dict(color='purple'),
    text=decade_stats['Songs'],
    textposition='outside',  # Place text outside the bars for better visibility
)

layout_songs_per_decade = go.Layout(
    title='Number of Songs Released Each Decade',
    xaxis=dict(title='Decade'),
    yaxis=dict(title='Number of Songs'),
    height=800,
)

fig_songs_per_decade = go.Figure(data=[data_songs_per_decade], layout=layout_songs_per_decade)

# Update layout for aesthetic improvements
fig_songs_per_decade.update_traces(textfont_size=16, textfont_color='black', insidetextanchor='start')  # Increase text font size and set color
fig_songs_per_decade.update_layout(plot_bgcolor='rgba(255, 255, 255, 0.9)',  # Lighter transparent plot background
                                   paper_bgcolor='rgba(255, 255, 255, 0.9)',  # Lighter transparent paper background
                                   font=dict(family="Arial, sans-serif", size=18, color='darkslategray'),  # Font settings
                                   title=dict(x=0.5, y=0.95, font=dict(size=28)),  # Title position and style
                                   xaxis=dict(tickfont=dict(size=14)),  # X-axis tick font size
                                   yaxis=dict(tickfont=dict(size=14)),  # Y-axis tick font size
                                   )

# Display plot
fig_songs_per_decade.show()
```

```python
import plotly.graph_objs as go

# Assuming 'song' DataFrame and necessary imports are already defined

# Count movies and songs per director
director_stats = song.groupby('Director').agg({'Movie': 'nunique', 'Songs': 'count'}).reset_index()

# Sort by number of songs in descending order
director_stats = director_stats.sort_values(by='Songs', ascending=False)

# Select top 15 directors by number of songs
top_directors = director_stats.head(15)

# Create a Plotly horizontal bar chart
data_director_stats = [
    go.Bar(
        y=top_directors['Director'],
        x=top_directors['Movie'],
        orientation='h',
        name='Movies',
        marker=dict(color='lightblue'),
        text=top_directors['Movie'],
        textposition='outside',  # Display text outside the bars
    ),
    go.Bar(
        y=top_directors['Director'],
        x=top_directors['Songs'],
        orientation='h',
        name='Songs',
        marker=dict(color='lightgreen'),
        text=top_directors['Songs'],
        textposition='outside',  # Display text outside the bars
    )
]

layout_director_stats = go.Layout(
    title='Top 15 Directors by Number of Movies and Songs',
    barmode='stack',
    xaxis=dict(title='Count'),
    yaxis=dict(title='Director'),
    height=800,
)

fig_director_stats = go.Figure(data=data_director_stats, layout=layout_director_stats)

# Update layout for aesthetic improvements
fig_director_stats.update_traces(textfont_size=14, textfont_color='black')  # Increase text font size and set color
fig_director_stats.update_layout(plot_bgcolor='rgba(255, 255, 255, 0.9)',  # Lighter transparent plot background
                                 paper_bgcolor='rgba(255, 255, 255, 0.9)',  # Lighter transparent paper background
                                 font=dict(family="Arial, sans-serif", size=18, color='darkslategray'),  # Font settings
                                 title=dict(x=0.5, y=0.95, font=dict(size=28)),  # Title position and style
                                 xaxis=dict(tickfont=dict(size=14)),  # X-axis tick font size
                                 yaxis=dict(tickfont=dict(size=14)),  # Y-axis tick font size
                                 )

# Display plot
fig_director_stats.show()
```

```python
import plotly.graph_objs as go

# Aggregate data by year
movies_songs_year = song.groupby('Year').agg({'Movie': 'nunique', 'Songs': 'count'}).reset_index()

# Create a scatter plot
fig_movies_songs_year = go.Figure(
    go.Scatter(
        x=movies_songs_year['Year'],
        y=movies_songs_year['Songs'],
        mode='lines+markers',
        marker=dict(color='purple', size=10),
        name='Songs',
        text=[f'Year: {year}<br>Songs: {songs}<br>Movies: {movies}' 
              for year, songs, movies in zip(movies_songs_year['Year'], 
                                             movies_songs_year['Songs'], 
                                             movies_songs_year['Movie'])],
        hoverinfo='text'
    )
)

fig_movies_songs_year.add_trace(
    go.Scatter(
        x=movies_songs_year['Year'],
        y=movies_songs_year['Movie'],
        mode='lines+markers',
        marker=dict(color='blue', size=10),
        name='Movies',
        text=[f'Year: {year}<br>Songs: {songs}<br>Movies: {movies}' 
              for year, songs, movies in zip(movies_songs_year['Year'], 
                                             movies_songs_year['Songs'], 
                                             movies_songs_year['Movie'])],
        hoverinfo='text'
    )
)

fig_movies_songs_year.update_layout(
    title='Number of Movies and Songs per Year',
    xaxis=dict(title='Year'),
    yaxis=dict(title='Count'),
    height=600,
    plot_bgcolor='rgba(255, 255, 255, 0.9)',  # Lighter transparent plot background
    paper_bgcolor='rgba(255, 255, 255, 0.9)',  # Lighter transparent paper background
    font=dict(family="Arial, sans-serif", size=14, color='darkslategray'),  # Font settings
)

fig_movies_songs_year.show()
```

```python
# Count number of movies per director and actor combination
movies_by_director_actor = song.groupby(['Director', 'Actor'])['Movie'].nunique().reset_index()

# Select top 10 combinations by number of movies
top_combinations = movies_by_director_actor.sort_values(by='Movie', ascending=False).head(10)

# Create a Plotly grouped bar chart
fig_movies_director_actor = go.Figure()

for actor, group in top_combinations.groupby('Actor'):
    fig_movies_director_actor.add_trace(
        go.Bar(
            x=group['Director'],
            y=group['Movie'],
            name=actor,
            text=group['Movie'],
            textposition='auto',
        )
    )

fig_movies_director_actor.update_layout(
    barmode='group',
    title='Number of Movies by Director and Actor (Top 10 Combinations)',
    xaxis=dict(title='Director'),
    yaxis=dict(title='Number of Movies'),
    height=800,
    plot_bgcolor='rgba(255, 255, 255, 0.9)',  # Lighter transparent plot background
    paper_bgcolor='rgba(255, 255, 255, 0.9)',  # Lighter transparent paper background
    font=dict(family="Arial, sans-serif", size=14, color='darkslategray'),  # Font settings
)

fig_movies_director_actor.show()
```

```python
# Count songs per year and actor
songs_per_year_actor = song.groupby(['Year', 'Actor'])['Songs'].count().reset_index()

# Select top 5 actors by total number of songs
top_actors = songs_per_year_actor.groupby('Actor')['Songs'].sum().nlargest(5).index

# Filter data for top actors
songs_per_year_actor_top = songs_per_year_actor[songs_per_year_actor['Actor'].isin(top_actors)]

# Create a Plotly line plot
fig_songs_over_time_actor = go.Figure()

for actor, data in songs_per_year_actor_top.groupby('Actor'):
    fig_songs_over_time_actor.add_trace(
        go.Scatter(
            x=data['Year'],
            y=data['Songs'],
            mode='lines+markers',
            name=actor,
            marker=dict(size=10),
            text=data['Songs'],
            hoverinfo='text',
        )
    )

fig_songs_over_time_actor.update_layout(
    title='Number of Songs Over Time by Actor (Top 5)',
    xaxis=dict(title='Year'),
    yaxis=dict(title='Number of Songs'),
    height=600,
    plot_bgcolor='rgba(255, 255, 255, 0.9)',  # Lighter transparent plot background
    paper_bgcolor='rgba(255, 255, 255, 0.9)',  # Lighter transparent paper background
    font=dict(family="Arial, sans-serif", size=14, color='darkslategray'),  # Font settings
)

fig_songs_over_time_actor.show()
```
